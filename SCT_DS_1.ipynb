```python
top_10countries=ds.sort_values(by='pop2023',ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x='pop2023',y='country',data=top_10countries,palette='plasma')
plt.title('top 10 countries by 2023 population')
plt.xlabel('population(in billions)')
plt.ylabel('countries')
plt.show()
```

```python
print(ds.columns)
```

```python
country_data=ds[ds['country']=='India']
years=['pop1980','pop2000','pop2010','pop2022','pop2023','pop2030','pop2050']
population=country_data[years].values.flatten()
plt.figure(figsize=(10,6))
plt.plot(years,population,marker='o',color='green')
plt.title('population growth in India')
plt.xlabel('year')
plt.ylabel('population in billions')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(ds['growthRate'],bins=15,kde=True,color='Blue')
plt.title('distribution of population growth rate')
plt.xlabel('growth rate(%)')
plt.ylabel('frequency')
plt.show()
```

```python
top5_countries=ds.sort_values(by='worldPercentage',ascending=False).head(5)
plt.figure(figsize=(10,10))
plt.pie(top5_countries['worldPercentage'],labels=top5_countries['country'],autopct='%1.1f%%',startangle=140,colors=sns.color_palette('pastel'))
plt.title('world population share by top5 countries')
plt.show()
```
